```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
import seaborn as sns
```

```python
TXN=pd.read_csv("ds_challenge_apr2021.csv")
```

```python
TXN.head()
```

### Exploratory Data Analysis

```python
TXN["TXN"]=1
```

```python
Fraudsters_Customers=pd.DataFrame(TXN[TXN["fraude"]==True].groupby("ID_USER")["TXN"].sum()).reset_index()
```

```python
Fraudsters_Customers["Is_Fraudster"]=1
```

```python
Line_Card=pd.DataFrame(TXN.groupby("ID_USER")["linea_tc"].apply(lambda x: pd.value_counts(x).index[0])).reset_index()
```

```python
Amount_By_Customer=pd.DataFrame(TXN.groupby("ID_USER")[["monto","TXN","dcto","cashback"]].sum()).reset_index()
```

```python
Customers=Amount_By_Customer.merge(Line_Card,on="ID_USER",how="left")
```

```python
Customers=Customers.merge(Fraudsters_Customers[["ID_USER","Is_Fraudster"]],on="ID_USER",how="left")
```

```python
Customers["Is_Fraudster"]=Customers["Is_Fraudster"].fillna(0)
```

```python
Customers.head(2)
```

```python
def density_plot(variable):
    plt.figure(figsize=(16,10))
    sns.set_style('whitegrid')
    sns.kdeplot(Customers[Customers["Is_Fraudster"]==1][variable], bw=0.5,label="Fraudsters")
    sns.kdeplot(Customers[Customers["Is_Fraudster"]==0][variable], bw=0.5,label="Good Users")
    plt.title(variable)
    plt.savefig(variable+'.png')
    plt.show()
```

```python
density_plot("monto")
```

```python
density_plot("TXN")
```

```python
density_plot("linea_tc")
```

```python
density_plot("cashback")
```

##### Extracting values of the "dispositivo" variable

```python
TXN["dispositivo"]=list(map(lambda x: ast.literal_eval(x),TXN["dispositivo"].values))
```

```python
Models=list()
Devices_scores=list()
Operative_Systems=list()
for i in range(0,len(TXN)):
    Models.append(TXN["dispositivo"][i]["model"])
    Devices_scores.append(TXN["dispositivo"][i]["device_score"])
    Operative_Systems.append(TXN["dispositivo"][i]["os"])
```

```python
TXN["model"]=Models
TXN["device_score"]=Devices_scores
TXN["os"]=Operative_Systems
```
